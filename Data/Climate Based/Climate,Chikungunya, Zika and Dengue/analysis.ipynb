```python
import numpy as np
import pandas as pd   
import seaborn as sns
import matplotlib.pyplot as plt
```

### Reading the CSV files 
We have 5 cities in which all the cities have the folllowing 

```python
manhuacu = pd.read_csv(r'data/Manhuacu.csv', skiprows=9,delimiter=';')  # Skip the first 10 rows (starting from line 11)
juiz_de_fora = pd.read_csv(r'data/Juiz_de_fora.csv', skiprows=9,delimiter=';') 
coronel_pacheco = pd.read_csv(r'data/Coronel_Pacheco.csv', skiprows=9,delimiter=';') 
muriae = pd.read_csv(r'data/Muriae.csv', skiprows=9,delimiter=';') 
```

```python
new_col_names = ['Measurement Date',
                 'TOTAL PRECIPITATION, DAILY (AUT)(mm)',
                 'AVERAGE DAILY ATMOSPHERIC PRESSURE (AUT)(mB)',
                 'AVERAGE DAILY DEW POINT TEMPERATURE (AUT)(°C)',
                 'MAXIMUM TEMPERATURE, DAILY (AUT)(°C)',
                 'AVERAGE TEMPERATURE, DAILY (AUT)(°C)',
                 'MINIMUM TEMPERATURE, DAILY (AUT)(°C)',
                 'RELATIVE AIR HUMIDITY, DAILY AVERAGE (AUT)(%)',
                 'RELATIVE AIR HUMIDITY, MINIMUM DAILY DAY (AUT)(%)',\
                 'WIND, MAXIMUM DAILY GUST (AUT)(m/s)',
                 'WIND, AVERAGE DAILY SPEED (AUT)(m/s)',
                 ]
manhuacu.drop(manhuacu.columns[-1],axis=1,inplace=True)
juiz_de_fora.drop(juiz_de_fora.columns[-1],axis=1,inplace=True)
coronel_pacheco.drop( coronel_pacheco.columns[-1],axis=1,inplace=True)
muriae.drop(muriae.columns[-1],axis=1,inplace=True)

manhuacu.columns = new_col_names
juiz_de_fora.columns = new_col_names
coronel_pacheco.columns = new_col_names
muriae.columns = new_col_names
```

```python
manhuacu.tail(10)
```

```python
numeric_manhuacu = manhuacu.select_dtypes(include=['float64'])
correlation_matrix =  numeric_manhuacu.corr()

# Create a heatmap using Seaborn
plt.figure(figsize=(10, 8))  # Adjust the figure size if needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap between Climatic Features')
plt.show()
```
